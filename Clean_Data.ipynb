## Clean Data Function

**Import necessary library**

```python
import pandas as pd
import numpy as np
```

**Main Function**

```python
def Clean_data(data_file):
    for i in range(len(data_file)):
        a = ''.join(filter(str.isdigit, data_file.loc[i, 'Temp']))
        data_file.loc[i, 'Temp'] = a
        if data_file.loc[i][3] == "No wind":
            data_file.loc[i, 'Wind'] = "0"
        else:
            b = ''.join(filter(str.isdigit, data_file.loc[i][3]))
            data_file.loc[i, 'Wind'] = b
        c = ''.join(filter(str.isdigit, data_file.loc[i][4]))
        data_file.loc[i, 'Humidity'] = c
        d = ''.join(filter(str.isdigit, data_file.loc[i][5]))
        data_file.loc[i, 'Barometer'] = d
        if type(data_file.loc[i, 'Visibility']) == str:
            e = ''.join(filter(str.isdigit, data_file.loc[i][6]))
            data_file.loc[i, 'Visibility'] = e
        else:
            data_file.loc[i, 'Visibility'] = data_file.loc[i, 'Visibility']
    return data_file
    
```

## Clean 1000 samples dataset

```python
data_file_1000 = pd.read_excel('data_1000_samples_raw.xlsx')
data_file_1000
```

```python
data_file_1000 = Clean_data(data_file_1000)
data_file_1000
```

```python
data_file_1000.to_excel('data_1000_samples_clean.xlsx', index=False)
```

# Standard Scaling
___

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
from matplotlib.projections.geo import Transform
import sklearn.preprocessing as sp
```

### Prepare Data

**1000 samples dataset**

```python
data_1000 = pd.read_excel("E:\\VSProject\\TTNT\\DataScience\\data_1000_samples_clean.xlsx").loc[:,['Temp', 'Wind', 'Humidity', 'Barometer']]
data_1000
```

**10000 samples dataset**

```python
data_10000 = pd.read_excel("E:\\VSProject\\TTNT\\DataScience\\data_10000_samples_clean.xlsx").loc[:,['Temp', 'Wind', 'Humidity', 'Barometer']]
data_10000
```

### Before standard scaling

```python
# Temp
plt.figure(figsize=(2,5))
sns.violinplot(data_1000['Temp'])

```

```python
# Humidity
plt.figure(figsize=(2,5))
ax = sns.violinplot(data_1000['Humidity'])
```

```python
# Wind
plt.figure(figsize=(2,5))
ax = sns.violinplot(data_1000['Wind'])
```

```python
# Barometer
plt.figure(figsize=(2,5))
ax = sns.violinplot(data_1000['Barometer'])
```

```python
# 4 feature on the same figure
plt.figure(figsize=(12, 20))
ax = sns.violinplot(data_10000)
_ = ax.set_xticklabels(data_10000.keys(), rotation=90)
```

* Features lies on variety of significantly different period (> 1000 for Barometer and <100 for the remaining)

### After standard scaling

```python
# Scaling for 1000 samples
data_st_1000 = pd.DataFrame(sp.StandardScaler().fit_transform(data_1000),columns= data_1000.columns) 
```

```python
# Scaling for 10000 samples
data_st_10000 = pd.DataFrame(sp.StandardScaler().fit_transform(data_10000),columns= data_10000.columns) 
```

```python
# 4 feature on the same figure (1000 samples)
plt.figure(figsize=(12, 6))
ax = sns.violinplot(data_st_1000)
_ = ax.set_xticklabels(data_st_1000.keys(), rotation=90)
```

```python
# 4 feature on the same figure (10000 samples)
plt.figure(figsize=(12, 6))
ax = sns.violinplot(data_st_10000)
_ = ax.set_xticklabels(data_st_10000.keys(), rotation=90)
```

* Features were scaled to standard distribution with mean = 0 and std = 1
* Easily know and balance compared between all features

### Save 1000 and 10000 samples dataset as excel file

```python
# 1000 samples
data_st_1000.to_excel('data_1000_samples_standard.xlsx', index = False)
```

```python
# 10000 samples
data_st_10000.to_excel('data_10000_samples_standard.xlsx', index = False)
```
